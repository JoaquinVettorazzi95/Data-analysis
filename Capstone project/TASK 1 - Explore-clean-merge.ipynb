```python
#TRANSPONEMOS LAS COLUMNAS DE DIAS Y HACEMOS QUE CADA DIA SEA UNA FILA.

df_sales = pd.melt(df_sales, id_vars=['id', 'item', 'category', 'department', 'store', 'store_code', 'region'],
                    var_name='d', value_name='quantity')
```

```python
df_cal = df_cal.drop(columns=['weekday_int'])
df_sales = df_sales.drop(columns=['id'])
```

**PRIMER MERGE**

```python
merged_sales_date = df_sales.merge(df_cal, on=['d'], how='left')
```

```python
merged_sales_date.head()
```

```python
merged_sales_date.shape
```

**eliminamos columnas que no nos sirven por ahora**

```python
merged_sales_date.drop('event', axis=1, inplace=True)
merged_sales_date.drop('weekday', axis=1, inplace=True)
merged_sales_date.drop('department', axis=1, inplace=True)
```

```python
merged_sales_date.tail()
```

```python
merged_sales_date['yearweek'] = merged_sales_date['date'].copy()
```

```python
merged_sales_date
```

```python
# Conviertimos la columna 'yearweek' a tipo datetime
merged_sales_date['yearweek'] = pd.to_datetime(merged_sales_date['yearweek'])
```

```python
# Obtenemos el año y la semana usando el método 'isocalendar()'
merged_sales_date['year'] = merged_sales_date['yearweek'].dt.isocalendar().year
merged_sales_date['week'] = merged_sales_date['yearweek'].dt.isocalendar().week
```

```python

# Combinamos el año y la semana en una nueva columna 'yearweek_numeric'
merged_sales_date['yearweek_numeric'] = merged_sales_date['year'] * 100 + merged_sales_date['week']
```

```python
merged_sales_date
```

```python
merged_sales_date.info()
```

```python
merged_sales_date.rename(columns={'yearweek_numeric': 'yearweek'}, inplace=True)
```

```python
merged_sales_date.drop('year', axis=1, inplace=True)
merged_sales_date.drop('week', axis=1, inplace=True)
merged_sales_date.drop('yearweek', axis=1, inplace=True)
```

```python
merged_sales_date.head()
```

```python
merged_sales_date.drop('d', axis=1, inplace=True)
```
